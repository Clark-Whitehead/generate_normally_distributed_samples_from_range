```python
import numpy as np
from scipy.stats import norm, truncnorm
import matplotlib.pyplot as plt
from collections import Counter
```

```python
def norm_distributed_samples(low,high,num_samples,plot=False):
    
    x = list(np.arange(low, high, (high-low)/num_samples))
    
    mean = np.mean(x)
    std = np.std(x)
    
    bottom, top = (low - mean) / std, (high - mean) / std
    
    probabilities = truncnorm.pdf(x, bottom, top, mean, std)
    
    normalized_probs = probabilities / probabilities.sum()
    
    output = np.random.choice(x, num_samples, p=normalized_probs)
    
    if plot == True:
    
        plt.figure(1, figsize=(16, 6))

        plt.subplot(1,2,1)
        plt.hist(output)

        plt.subplot(1,2,2)
        plt.plot(probabilities)

        plt.show()
    
    return output

norm_distributed_samples(5, 20, 1000, plot=True)
```
